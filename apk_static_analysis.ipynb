```python
import pandas as pd

drebin_df = pd.read_csv("/Users/mac/Desktop/apk/apk_ml_model/CSV/shuffled_apk_dataset.csv", low_memory=False)
malgenome_df = pd.read_csv("/Users/mac/Desktop/apk/apk_ml_model/CSV/malgenome-215_shuffled.csv", low_memory=False)

```

```python
# checking the number of rows and column in each dataset
print("Drebin columns:", len(drebin_df.columns))
print("Malgenome columns:", len(malgenome_df.columns))

print("Drebin column names:\n", drebin_df.columns)
print("Malgenome column names:\n", malgenome_df.columns)
```

```python


# Converting the  columns to sets
drebin_cols = set(drebin_df.columns)
malgenome_cols = set(malgenome_df.columns)

# Checking the  differences
only_in_drebin = drebin_cols - malgenome_cols
only_in_malgenome = malgenome_cols - drebin_cols

print("Features only in Drebin:", only_in_drebin)
print("Features only in Malgenome:", only_in_malgenome)

# Check if columns are exactly the same
if drebin_cols == malgenome_cols:
    print("✅ Both datasets have the same features.")
else:
    print("❌ Datasets have different features.")
```
